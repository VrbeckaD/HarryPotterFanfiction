```python
matrix = data["all"].to_numpy()
names = matrix[:,0]
items = matrix[:,1:]
```

```python
max_values = []
for rn, row in enumerate(items):
    for cn, value in enumerate(row):
        if rn >= cn:
            max_values.append((value, "/".join(sorted([names[cn], names[rn]]))))
max_values.sort(key=lambda rec: -rec[0])
pd.DataFrame(max_values[:100])
```
