```python
import math

import pandas as pd
import holoviews as hv
from holoviews import dim
import numpy


hv.extension('matplotlib')
hv.output(fig='svg', size=300)
```

https://holoviews.org/reference/elements/bokeh/Chord.html

```python
data = pd.read_csv("lovers_matrix_slash.csv")
data
```

```python
matrix = data.to_numpy()
keys = tuple(matrix[:, 0])
index2key = dict(enumerate(keys))
key2index = {k: i for i, k in enumerate(keys)}


def log(val):
    if val < 10:
        return 0
    elif val < 50:
        return 5
    elif val < 100:
        return 10
    elif val < 200:
        return 15
    elif val < 500:
        return 20
    elif val < 1000:
        return 25
    elif val < 2000:
        return 30
    else:
        return 60

def node_name(original):
    index = key2index[original]
    total = sum(int(i) for i in matrix[index, 1:])
    if total < 10:
        return "others"
    if total < 105: 
        return ""
    return original.replace("_", " ")



routes = []
for r, row in enumerate(matrix):
    key = row[0]
    for c, col in enumerate(row[1:], start=1):
        if r >= c - 1:
            if int(matrix[r, c]) >50:
                routes.append((key2index[key], c - 1, log(int(matrix[r, c]))))
routes = pd.DataFrame.from_records(data=routes, columns=["source", "target", "value"])

new_nodes = hv.core.data.Dataset(pd.DataFrame.from_records([(key2index[key], node_name(key), key) for key in keys], columns=["index", "name", "group"]), 'index')

routes
```

```python
def rotate_label(plot, element):    
    labels = plot.handles["labels"]
    for annotation in labels:        
        angle = annotation.get_rotation()
        if 90 < angle < 270:
            annotation.set_rotation(180 + angle)
            annotation.set_horizontalalignment("right")
  
```

V tomto seku je oeten rotace popisk okolo kruhu. Prav polovina (1. a 4. kvadrt) x lev polovina (2. a 3. kvadrt).

```python
graph = hv.Chord((routes, new_nodes))
```

```python
graph.opts(cmap='Category20',
           edge_cmap='Category20', 
           edge_color=dim('source').str(), 
           labels='name', 
           node_color=dim('index').str(),
           hooks=[rotate_label]
           )
```
